```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

def get_top_bigrams_overall(df, text_column='normalized', top_n=10):
    """
    Fungsi untuk mendapatkan top bigram dari seluruh dataset tanpa memisahkan berdasarkan sentimen.
    """
    if df.empty:
        return []

    vectorizer = CountVectorizer(ngram_range=(2,2))
    X = vectorizer.fit_transform(df[text_column])

    bigram_freq = zip(vectorizer.get_feature_names_out(), X.sum(axis=0).tolist()[0])
    bigram_freq = sorted(bigram_freq, key=lambda x: x[1], reverse=True)[:top_n]

    return bigram_freq

def bigram_to_df_overall(bigram_list):
    """
    Fungsi untuk mengkonversi list bigram menjadi DataFrame.
    """
    return pd.DataFrame(bigram_list, columns=["Bigram", "Frequency"])

bigram_mrt_overall = get_top_bigrams_overall(original_df_mrt, text_column='stemmed', top_n=10)

bigram_krl_overall = get_top_bigrams_overall(original_df_krl, text_column='normalized', top_n=10)

df_bigram_mrt_overall = bigram_to_df_overall(bigram_mrt_overall)
df_bigram_krl_overall = bigram_to_df_overall(bigram_krl_overall)

print("Top Bigrams for MRT (Overall):")
print(df_bigram_mrt_overall)

# # Simpan ke CSV
# df_bigram_mrt_overall.to_csv('bigram_mrt_overall.csv', index=False)
# df_bigram_krl_overall.to_csv('bigram_krl_overall.csv', index=False)
```

```python
def get_top_bigrams(df, label_column='label_id', text_column='normalized', top_n=10):
    results = {}

    sentiment_map = {
        0: "Negative",
        1: "Neutral",
        2: "Positive"
    }

    for label_id, sentiment_name in sentiment_map.items():
        subset = df[df[label_column] == sentiment_name]

        if subset.empty:
            results[sentiment_name] = []
            continue

        vectorizer = CountVectorizer(ngram_range=(2,2))
        X = vectorizer.fit_transform(subset[text_column])

        bigram_freq = zip(vectorizer.get_feature_names_out(), X.sum(axis=0).tolist()[0])
        bigram_freq = sorted(bigram_freq, key=lambda x: x[1], reverse=True)[:top_n]

        results[sentiment_name] = bigram_freq

    return results
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

def get_top_bigrams_overall(df, text_column='normalized', top_n=10):
    """
    Fungsi untuk mendapatkan top bigram dari seluruh dataset tanpa memisahkan berdasarkan sentimen.
    """
    if df.empty:
        return []

    vectorizer = CountVectorizer(ngram_range=(2,2))   # BIGRAM
    X = vectorizer.fit_transform(df[text_column])

    bigram_freq = zip(vectorizer.get_feature_names_out(), X.sum(axis=0).tolist()[0])
    bigram_freq = sorted(bigram_freq, key=lambda x: x[1], reverse=True)[:top_n]

    return bigram_freq
```

```python
def get_example_texts_for_bigrams(df, bigram_list, text_column='full_text', n=5, random_sample=True):
    examples = {}
    for bigram, _ in bigram_list:
        # Filter teks yang mengandung bigram
        subset = df[df[text_column].str.contains(bigram, case=True, na=True)]
        if random_sample:
            # Ambil sample random sebanyak n, atau semua jika kurang dari n
            subset = subset.sample(n=min(n, len(subset)), random_state=None)  # random_state=None untuk benar-benar random
        else:
            # Ambil dari atas (deterministic)
            subset = subset.head(n)
        examples[bigram] = subset[text_column].tolist()
    return examples
```

```python
def examples_to_df(examples_dict, source_name):
    rows = []
    for bigram, tweets in examples_dict.items():
        for tweet in tweets:
            rows.append({"Bigram": bigram, "Tweet": tweet, "Source": source_name})
    return pd.DataFrame(rows)
```

```python
bigram_mrt_overall = get_top_bigrams_overall(original_df_mrt, text_column='stemmed', top_n=10)
examples_mrt = get_example_texts_for_bigrams(original_df_mrt, bigram_mrt_overall, text_column='stemmed', n=1)
df_examples_mrt = examples_to_df(examples_mrt, 'MRT')
df_examples_mrt
# Menyimpan
# df_examples_mrt.to_csv('examples_mrt.csv', index=False)
```

```python
def get_example_tweets(df, label_column='label_id', text_column='full_text', n=5):
    sentiment_map = {
        0: "Negative",
        1: "Neutral",
        2: "Positive"
    }

    examples = {}

    for label_id, sentiment_name in sentiment_map.items():
        subset = df[df[label_column] == sentiment_name][text_column].head(n).tolist()
        examples[sentiment_name] = subset

    return examples
```
